# **Question 2**

(30 points). Write a python program to **clean the text data** you collected above and save the data in a new column in the csv file. The data cleaning steps include:

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the [stopwords list](https://gist.github.com/sebleier/554280).

(4) Lowercase all texts

(5) Stemming. 

(6) Lemmatization.

```python
# Write your code here
#Write a python program to clean the text data you collected above and save the data in a new column in the csv file. The data cleaning steps include:
import pandas as pd
data_url = "tweets.csv"
train = pd.read_csv(data_url)
train['word_count'] = train['text'].apply(lambda x: len(str(x).split(" ")))

#1.remove noise such as special characters
import re
train['text'] = train['text'].apply(lambda x: re.sub('[^A-Za-z0-9\s]+', '', x))

#punctuation
train['text'] = train['text'].str.replace('[^\w\s]','')

#(2) Remove numbers.
train['text'] = train['text'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))

#(3) Remove stopwords by using the stopwords list.
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop = stopwords.words('english')
train['stopwords'] = train['text'].apply(lambda x: len([x for x in x.split() if x in stop]))
train[['text','stopwords']].head()
stop = stopwords.words('english')
train['text'] = train['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))


#(4) Lowercase all texts
train['text'] = train['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))

#(5) Stemming.
from nltk.stem import PorterStemmer
st = PorterStemmer()
train['text'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#(6) Lemmatization.
from textblob import Word
import nltk
nltk.download('wordnet')
nltk.download("omw-1.4")
train['text'] = train['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
train['text'].head()

print(train['text'].head())






```

**Write your explanations of the constituency parsing tree and dependency parsing tree here (Question 3-2):** 


constituency parsing tree :
The constituent parsing tree represents a sentence's hierarchical structure in terms of its fundamental elements, such as noun phrases, verb phrases, prepositional phrases, and so on. Each node in the tree represents a constituent part of the sentence, and its children are the constituent subparts. The tree's root node represents the entire sentence, whereas the leaf nodes represent individual words. Constituency parsing can be used for tasks such as sentence generation, machine translation, and summarization by identifying the grammatical structure of a sentence.

dependency parsing tree :
 The dependency parsing tree represents the grammatical roles and dependencies that exist between words in a sentence. Each node in the tree represents a word in the sentence, and the words on which it depends are represented by its children (i.e., the words that it modifies or is modified by). The root node of the tree represents the sentence's main verb or predicate, while the other nodes represent the subject, object, modifiers, and other verb dependents. Dependency parsing can be used for tasks such as information extraction, question answering, and sentiment analysis by identifying the semantic relationships between words in a sentence.
